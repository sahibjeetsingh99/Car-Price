```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor=RandomForestRegressor()
```

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
 #Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
random_regressor = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rsv_regressor_random = RandomizedSearchCV(estimator = random_regressor, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rsv_regressor_random.fit(X_train,y_train)
```

```python
rsv_regressor_random.best_params_
```

```python
rsv_regressor_random.best_score_
```

```python
y_pred=rsv_regressor_random.predict(X_test)
```

```python
sns.distplot(y_test-y_pred)
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn import metrics
```

```python
print('Mean absolute error:')
print( metrics.mean_absolute_error(y_test, y_pred))
```

```python
print('Mean Squared Error:')
print(metrics.mean_squared_error(y_test, y_pred))
```

```python
print('Root Mean Squared Error:')
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
